```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
import numpy as np
import scipy.stats as st
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import precision_recall_curve
```

```python
def histbox(x, col):
    
    fig, ax = pyplot.subplots(1, 2, figsize=(12, 5))
    sns.histplot(ax=ax[0], x=x, kde=True, bins=30, color=col)
    ax[0].set_ylabel('Частота')
    ax[0].set_title(f'Распределение показателей {x.name}', fontsize=12)
    sns.boxplot(ax=ax[1], x=x, color=col, medianprops={'color': "darkred", "linewidth": 2})
    ax[1].set_title(f'Диаграмма размаха {x.name}', fontsize = 12)
```

```python
def matrix_estimated (target, predict):
    
    cm = confusion_matrix(target, predict)
    sns.heatmap(cm, annot=True, cmap='Blues')
    plt.gca().set(xlabel = 'Predict', ylabel='Test')
    plt.title('Матрица типичных ошибок', fontsize=12)
    
    print('precision: ', precision_score(target, predict))
    print('recall: ', recall_score(target, predict))
```

```python
def predobr(data, t, num, cat):
    X = data.drop(t, axis=1)
    y = data[t]

    RANDOM_STATE = 42

    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=RANDOM_STATE)

    cat_col = cat
    num_col = num

    scaler = StandardScaler()
    encoder = OneHotEncoder(drop='first', sparse=False)

    X_train_scaled = scaler.fit_transform(X_train[num_col])
    X_test_scaled = scaler.transform(X_test[num_col])

    X_train_ohe = encoder.fit_transform(X_train[cat_col])
    X_test_ohe = encoder.transform(X_test[cat_col])

    X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col)
    X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col)

    X_train_ohe = pd.DataFrame(X_train_ohe, columns=cat_col)
    X_test_ohe = pd.DataFrame(X_test_ohe, columns=cat_col)

    X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
    X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
    
    return X_train, X_test, y_train, y_test  
```

```python
def linregfunc(X_train, X_test, y_train, y_test):
    
    lr = LinearRegression()
    lr = lr.fit(X_train, y_train)
    predictions = lr.predict(X_test)
    residuals = y_test - predictions
    
    return y_test, predictions, residuals
```

```python
def metrics(y_test, predictions, residuals):

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
    axes[0].hist(x=residuals, bins=30)
    axes[0].axvline(x=residuals.mean(), color='red')
    axes[0].set_title('Гистограмма распределения остатков')
    axes[0].set_xlabel('Остатки')

    axes[1].scatter(x=predictions, y=residuals)
    axes[1].set_xlabel('Предсказания модели')
    axes[1].set_ylabel('Остатки')
    axes[1].set_title('Анализ дисперсии')
    
    print('MSE: ', mean_squared_error(y_test, predictions))
    print('RMSE: ', mean_squared_error(y_test, predictions, squared=False))
    print('MAE: ', mean_absolute_error(y_test, predictions))
    print('R2: ', r2_score(y_test, predictions))
```
